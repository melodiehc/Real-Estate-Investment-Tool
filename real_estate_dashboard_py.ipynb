```python
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt
from math import exp

# Helper functions for calculations

def down_payment(price, percent):

    #Calculate the down payment for a property.
    Args:
        price (float) #Total price of the property.
        percent (float) #Percentage of the price to be paid upfront as the down payment.
    Returns:
        float #Down payment amount.

    return price * (percent / 100)

def mortgage_monthly(price, years, rate):

    Calculate the monthly mortgage payment using the loan amount, interest rate, and term in years.
    Args:
        price (float) #Total price of the property.
        years (int) #Loan term in years.
        rate (float) #Annual interest rate as a percentage.
    Returns:
        float #Monthly mortgage payment.

    percent = rate / 100
    down = down_payment(price, 20)  # Assuming a 20% down payment
    loan = price - down  # Loan amount after down payment
    months = years * 12  # Total number of payments
    monthly_rate = percent / 12  # Monthly interest rate
    # Monthly payment formula for fixed-rate mortgage
    payment = loan * monthly_rate / (1 - (1 + monthly_rate) ** (-months))
    return payment

def cap_rate(monthly_income, price):

    Calculate the capitalization rate (cap rate).
    Args:
        monthly_income (float) #Net income per month.
        price (float) #Total property price.
    Returns:
        float #Cap rate as a percentage.

    return (monthly_income * 12 / price) * 100

def cash_on_cash(monthly_income, down_payment):

    #Calculate the cash-on-cash return rate.
    Args:
        monthly_income (float) #Net income per month.
        down_payment (float) #Down payment made for the property.
    Returns:
        float  #Cash-on-cash return rate as a percentage.

    return (monthly_income * 12 / down_payment) * 100

# Streamlit dashboard setup
st.set_page_config(page_title="Real Estate Toolkit", layout="wide")  # Configure the page layout and title

st.title("Real Estate Investment Dashboard")  # Display the main title of the app

# Sidebar Inputs
st.sidebar.header("Property Details")  # Sidebar section for property details
price = st.sidebar.number_input("Property Price ($)", min_value=10000, step=1000, value=300000)  # Property price input
rent = st.sidebar.number_input("Monthly Rent ($)", min_value=500, step=50, value=2000)  # Monthly rent input
tax_rate = st.sidebar.number_input("Annual Tax Rate (%)", min_value=0.0, step=0.1, value=1.0)  # Annual property tax rate
loan_term = st.sidebar.slider("Loan Term (Years)", min_value=5, max_value=30, value=20, step=1)  # Loan term slider
interest_rate = st.sidebar.number_input("Annual Interest Rate (%)", min_value=0.0, step=0.1, value=4.0)  # Annual interest rate

# Sidebar section for additional expenses
st.sidebar.header("Additional Expenses")
management_fee = st.sidebar.number_input("Management Fee (% of Rent)", min_value=0.0, step=0.1, value=10.0)  # Management fee
repair_cost = st.sidebar.number_input("Repair Cost (% of Property Price)", min_value=0.0, step=0.1, value=2.0)  # Repair costs
vacancy_rate = st.sidebar.number_input("Vacancy Rate (% of Rent)", min_value=0.0, step=0.1, value=2.0)  # Vacancy rate

# Calculations for financial metrics
monthly_mortgage = mortgage_monthly(price, loan_term, interest_rate)  # Calculate monthly mortgage payment
property_tax = (price * (tax_rate / 100)) / 12  # Calculate monthly property tax
management_cost = rent * (management_fee / 100)  # Calculate monthly management cost
repair_cost_monthly = (price * (repair_cost / 100)) / 12  # Calculate monthly repair allowance
vacancy_loss = rent * (vacancy_rate / 100)  # Calculate monthly loss due to vacancies

# Net income calculations
monthly_net_income = rent - (monthly_mortgage + property_tax + management_cost + repair_cost_monthly + vacancy_loss)
annual_net_income = monthly_net_income * 12

# Financial metrics
cap_rate_value = cap_rate(monthly_net_income, price)  # Capitalization rate
down_payment_value = down_payment(price, 20)  # Down payment (20% assumed)
cash_on_cash_value = cash_on_cash(monthly_net_income, down_payment_value)  # Cash-on-cash return

# Main Dashboard
st.header("Property Performance Metrics")
st.write("### Key Metrics")
metrics = {
    "Monthly Cash Flow ($)": monthly_net_income,
    "Annual Net Income ($)": annual_net_income,
    "Cap Rate (%)": cap_rate_value,
    "Down Payment ($)": down_payment_value,
    "Cash-on-Cash Return (%)": cash_on_cash_value,
}
metrics_df = pd.DataFrame(metrics, index=["Value"])
st.dataframe(metrics_df.T)  # Display metrics in a table format

# Expense Breakdown Chart
st.header("Expense Breakdown")
expenses = {
    "Mortgage Payment": monthly_mortgage,
    "Property Tax": property_tax,
    "Management Cost": management_cost,
    "Repair Allowance": repair_cost_monthly,
    "Vacancy Loss": vacancy_loss,
}
expenses_df = pd.DataFrame.from_dict(expenses, orient="index", columns=["Amount"])
st.bar_chart(expenses_df)  # Visualize expenses as a bar chart

# Portfolio Simulation Section
st.header("Portfolio Simulation")
st.write("### Simulate the Performance of Multiple Properties")
portfolio_properties = st.number_input("Number of Properties", min_value=1, step=1, value=1)  # Number of properties input
total_income = annual_net_income * portfolio_properties  # Total annual income for all properties
total_investment = down_payment_value * portfolio_properties  # Total investment for all properties

# Display portfolio metrics
st.write(f"**Total Annual Income for {portfolio_properties} Properties:** ${total_income:,.2f}")
st.write(f"**Total Investment for {portfolio_properties} Properties:** ${total_investment:,.2f}")

# Summary
st.success("Dashboard Loaded Successfully! Use the sidebar to adjust parameters and explore different scenarios.")
```
